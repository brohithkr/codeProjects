```python
rand = np.random.RandomState(342453)
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
cars = rand.choice(np.arange(10000,65000,5000),12)
aeroplane = [0]*3 + [3] + [0]*3 + [2] + [0]*3 + [4]
cars
```

```python
df = pd.DataFrame(
    {
        "Cars":cars,
        "Aeroplane":aeroplane
    },
    index=months
)
df
```

```python
plt.scatter(df.index,df["Cars"],marker="D")
plt.scatter(df.index,df["Aeroplane"],marker="s")
plt.legend(["Car Sales","Aeroplane Sales"])
plt.title("Sale Data")
# plt.
# plt.scatter()
# plt.figure(figsize=(5,5))
# plt.savefig("hello.svg")
plt.show()
```
