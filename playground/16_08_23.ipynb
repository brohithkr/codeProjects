```python
import numpy as np
import pandas as pd
```

```python
rand = np.random.RandomState(23423343)
names = ["Nita","Vinya","Suresh","Rohith","Neerja"]
ages = rand.randint(12,25,len(names))
math_data = rand.randint(60,95,len(names))
math_data = math_data.astype(dtype=np.float32)
hobbies = rand.choice(["Reading","Gaming","Cycling"],len(names))
rollno = np.arange(0,len(names))
rollno = rollno.astype(np.str_)
# rollno[:] = "22BD1A05" + rollno[:]
rollno = np.array(list(map(lambda x: "22BD1A050" + x,rollno)))
df = pd.DataFrame({
    "name":names,
    "age":ages,
    "hobby":hobbies,
    # "math":math_data
},index=rollno)
df
```

```python
df2 = pd.DataFrame(
    {
        "hobby":["Reading","Gaming","Cycling"],
        "group":["A","B","C"]
    }
)
pd.merge(df,df2,on="hobby",how="left")
```

```python
data = pd.DataFrame(
    {
        "name":["Priya","Ram","Raj"],
        "math": [37,49,50],
        "eng": [48,48,48],
        "sci":[2,3,-10]
    }
)

print(data)

plt.plot(["math","eng","sci"],data.loc[0][data.columns != "name"],)
plt.plot(["math","eng","sci"],data.loc[2][data.columns != "name"],)
plt.plot(["math","eng","sci"],data.loc[1][data.columns != "name"],)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend(data["name"],loc="upper right")
plt.show()
```
