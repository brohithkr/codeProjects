```python
plt.plot([1,2,3,4],[3,4,5,6])
plt.errorbar(
    x = [1,2],
    y = [3,4],
    yerr=[.3,.1],
    xerr=[.2,.1],
    fmt='o'
)
plt.show()
```

```python
rand = np.random.RandomState(283728)
x = rand.randint(20,60,5)
y = rand.randint(40,80,5)

plt.scatter(x,y)

plt.errorbar(
    x = x[y<60],
    y = y[y<60],
    xerr = 1,
    yerr = 2,
    fmt = 'o',
    color = 'red'
)
print(x,y)
plt.show()
```
